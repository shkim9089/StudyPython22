###국적코드표 데이터 읽어 들이기

```python
#sample codemaster.xlsx 읽어들인후 조회하기
#데이터 프레임변수 code_master
```

```python
code_master=pd.read_excel('./files/sample_codemaster.xlsx')
```

```python
code_master
```

```python
sample_1.merge(code_master,left_on='국적코드',right_on='국적코드')
```

```python
pd.merge(sample_1,code_master)
```

```python
pd.merge(left=sample_1,
        right=code_master,
        how ='inner',
        left_on='국적코드',
        right_on='국적코드')
```

```python
pd.merge(left=sample_1,
        right=code_master,
        how ='left',
        left_on='국적코드',
        right_on='국적코드')  ##leff면은 없으면 없는데로 가져와라
```

```python
pd.merge(left=sample_1,
        right=code_master,
        how ='outer',
        left_on='국적코드',
        right_on='국적코드')  ##양쪽에다 없는데로 가져옴 ㅇㅇ
```
