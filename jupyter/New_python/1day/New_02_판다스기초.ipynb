### 국적코드표 데이터 읽어 들이기

```python
# sample_codemaster.xlsx 파일 읽어들인 후 조히하기
# 데이터프레임변수 : code_master
file_path = "./files/sample_codemaster.xlsx"
code_master = pd.read_excel(file_path)
code_master
```

```python
# 원본에 국적명 통합하기...merge()
pd.merge(sample_1, code_master)
```

```python
pd.merge(left = sample_1,
          right = code_master,
          how = "inner",
          left_on = "국적코드",
          right_on = "국적코드")
```

```python
pd.merge(left = sample_1,
          right = code_master,
          how = "left",
          left_on = "국적코드",
          right_on = "국적코드")
```

```python
# 데이터프레임변수명 : sample_2
# file : sample_2.xlsx
file_path = "./files/sample_2.xlsx"
sample_2 = pd.read_excel(file_path,
                        header = 1,
                        skipfooter = 2,
                        usecols = "A:C")
sample_2
```
