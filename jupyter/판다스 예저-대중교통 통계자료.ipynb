```python
import pandas as pd
```

```python
df = pd.read_csv('./data/subway_202101.csv', thousands=',')
df.head()
```

```python
df['무임승차비율'] = df['무임승차']  / (df['무임승차'] + df['유임승차'])
```

```python
df[df['무임승차비율'] == df['무임승차비율'].max()]
```

```python
station = df[df['무임승차비율'] == df['무임승차비율'].max()].iloc[0,4:8]
유임승차 = station['유임승차']
유임하차 = station['유임하차']
무임승차 = station['무임승차']
무임하차 = station['무임하차']
total = station.sum(axis=0)
```

```python
total, 유임승차, 유임하차, 무임승차, 무임하차
```

```python
import matplotlib.pyplot as plt
```

```python
plt.rc('font', family = 'Malgun Gothic')
```

```python
plt.pie(
[
    유임승차 / total * 100 , 유임하차 / total * 100, 무임승차 / total * 100, 
    무임하차 / total * 100
], labels = ['유임승차','유임하차','무임승차','무임하차'],
autopct = '%.1f%%', startangle=90)
```

```python
mx = df .iloc[:,range(4,52,2)].max(axis=0).tolist()  #tolist 리스트로 바꿔주는것
print(mx)
```

```python
mx_station =[]
for i , e in enumerate(mx):
    if e:
        mx_station.append('%s (%s)'
                          %(df[df.iloc[:, i*2+4]==e]["Unnamed: 3"].iloc[0],i+4))
```

```python
mx_station
```

```python
plt.bar(range(len(mx_station)),mx,color='blue')
plt.xticks(range(len(mx_station)),mx_station,rotation=90)
plt.show()
```
