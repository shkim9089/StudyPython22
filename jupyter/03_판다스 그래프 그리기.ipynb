```python
tips = sns.load_dataset("tips")
print(tips.head())
```

```python
fig = plt.figure()
axes1 = fig.add_subplot(1,1,1)
```

```python
axes1.hist(tips['total_bill'],bins=10,color='red')  #bins
axes1.set_title('Historgram od Total Bill',color='white')
axes1.set_xlabel('Frequency',color='white')
axes1.set_ylabel('Toatal Bill',color='white')
fig
```

```python
scatter_plot=plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(tips['total_bill'],tips['tip'],c='magenta')
axes1.set_title('Scatterplot of Total Bill vs Tip',color='white')
axes1.set_xlabel('Total Bill',color='white')
axes1.set_ylabel('Tips',color='white')
```

```python
boxplot = plt.figure()
axe1 = boxplot.add_subplot(1,1,1)
axes1.boxplot([tips[tips['sex']=='Female']['tip'],
               tips[tips['sex']=='Male']['tip']],
                    labels=['Female','Male'])
axes1.set_xlabel('Sex')
axes1.set_ylabel('Tip')
axe1.set_title('Boxplot of TipS by Sex')
```

```python
## 
```

```python
def recode_gender(gender):
    if gender =='Female':
        return 'red'
    else:
        return 'blue'

tips['color']=tips['sex'].apply(recode_gender)
tips.head()
```

```python
scatter_plot=plt.figure()
axes1 = scatter_plot.add_subplot(1,1,1)
axes1.scatter(
    x=tips['total_bill'],
    y=tips['tip'],
    s=tips['size']*10,   # 
    c=tips['color'],
    alpha=0.5)

axes1.set_title('S Total Bill vs Tip colored by sex and sized by size',color='white')
axes1.set_xlabel('Total Bill',color='white')
axes1.set_ylabel('Tips',color='white')
```

```python
## 
# 

bar_plot = plt.figure()
axes1=bar_plot.add_subplot(1,1,1)

axes1.bar(tips['sex'],tips['total_bill'].count())
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'])
ax.set_title('Total Bill Histogram with Density plot')
```

```python
ax = plt.subplots()
ax = sns.distplot(tips['total_bill'],kde=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplot()
ax = sns.regplot(x='total_bill',y='tip',data=tips)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
ax = plt.subplot()
ax = sns.regplot(x='total_bill',y='tip',data=tips,fit_reg=False)
ax.set_title('Total Bill Histogram')
ax.set_xlabel('Total Bill')
ax.set_ylabel('Frequency')
```

```python
joint =sns.jointplot(x='total_bill',y='tip',data=tips)
joint.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total bill and Tip',fontsize=10,y=1.03)
```

```python
joint =sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
joint.set_axis_labels(xlabel='Total Bill',ylabel='Tip')
joint.fig.suptitle('Joint Plot of Total bill and Tip',fontsize=10,y=1.03)
```

```python
ax = plt.subplot()
ax =sns.kdeplot(data=tips['total_bill'],
               data2=tips['tip'],
               shade=True)
ax.set_title('kernel desitty plot of totla bull and tip')
ax.set_xlabel('total bill')
ax.set_ylabel('tip')
```

```python
ax =plt.subplot()
ax= sns.boxplot(x='time',y='total_bill',data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('time of day')
ax.set_ylabel('totla bill')
```

```python
ax =plt.subplot()
ax = sns.violinplot(x='time',y='total_bill',data=tips)
ax.set_title('Boxplot of total bill by time of day')
ax.set_xlabel('time of day')
ax.set_ylabel('totla bill')
```
